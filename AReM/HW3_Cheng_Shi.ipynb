(c) ii
Extract the time-domain features minimum, maximum, mean, median, standard deviation, first quartile, and third quartile for all of the 6 time series in each instance. You are free to normalize/standardize features or use them

```python
temp_nested_list = []
paths_csv = []
for root,  dirs, files in os.walk(os.getcwd()):
    for f in files:
        fullpath = os.path.join(root,f)
        if not f.startswith('.') and f.endswith('.csv'):
            end = fullpath.find('dataset')          
            start = fullpath.find('AReM/')
            label = fullpath[start+5: end-1]
            paths_csv.append([fullpath,label])
            #path will be paths_csv[i][0], label will be paths_csv[i][1]
            
feature_cols = ['avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
for i in feature_cols:
    min_n = dict()
    max_n = dict()
    mean_n = dict()
    med_n = dict()
    std_n = dict()
    q1_n = dict()
    q3_n = dict()
    label_n = dict()

for col in feature_cols:
    min_n [col] = list()
    max_n [col]= list()
    mean_n[col] = list()
    med_n[col] = list()
    std_n[col] = list()
    q1_n[col]= list()
    q3_n[col] = list()
    label_n[col] = list()

for index in range(0,len(paths_csv)):
    df = pd.read_csv(paths_csv[index][0],skiprows = 5, header = None, names= ['time','avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23'])
    df['label'] = paths_csv[index][1]
    #print(df)
    #df = df_tem.iloc[: , 1:]
    
    
    for i in feature_cols:
        #if i==feature_cols[3] and index==57: 
        #print('Hello', index, paths_csv[index], '\n')
        min_n[i].append(np.nanmin(df[i]))
        max_n[i].append(np.nanmax(df[i]))
        mean_n[i].append(np.nanmean(df[i]))
        med_n[i].append(np.nanmedian(df[i]))
        std_n[i].append(np.nanstd(df[i]))
        q1_n[i].append(np.nanquantile(df[i],0.25))
        q3_n[i].append(np.nanquantile(df[i],0.75))
        label_n[i].append(df['label'][0])
#print(len(label_n['avg_rss12']))
        
```

```python
list_dict = [min_n, max_n, mean_n, med_n, std_n, q1_n, q3_n,label_n]
time_domain_name = ['min','max','mean','med','std','q1','q3','label']
#list_dict[0]['avg_rss12']
#list_dict[0].keys()


schema = []
num_records = len (min_n[feature_cols[0]])
list_2d_temp = list () 
for record_index in range (0, num_records):
    record = list () 
    for col_index in range (1, len (feature_cols) + 1):
        col = feature_cols [col_index-1]
        for metric_index in range (0, len (time_domain_name)):
            metric_name = time_domain_name [metric_index] + str (col_index)
            cell = list_dict [metric_index][col][record_index]
            record.append (cell)
            if metric_name not in schema: schema.append (metric_name)
    list_2d_temp.append (record)

#print (schema)
#print(list_2d_temp[0])


feature_desc = pd.DataFrame(list_2d_temp, columns = schema)
feature_desc
```
